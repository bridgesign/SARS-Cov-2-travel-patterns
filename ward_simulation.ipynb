```python
# Extracting a random 5000, 10000 and 20000 population patch
def create_random_patch(G):
    patch_5 = []

    pop = 0

    node_list = list(G.nodes())

    point = np.random.choice(node_list)

    while pop<5000:
        pop+=G.nodes[point]['NP']
        patch_5.append((G.nodes[point]['Y'],G.nodes[point]['X']))
        node_list.remove(point)
        mdis = 999999999999999999999999999999
        for j in node_list:
            dis = 0
            for k in patch_5:
                dis+=(k[0]-G.nodes[j]['Y'])**2 + (k[1]-G.nodes[j]['X'])**2
            if dis<mdis:
                point = j
                mdis = dis

    patch_10 = patch_5[:]

    while pop<10000:
        pop+=G.nodes[point]['NP']
        patch_10.append((G.nodes[point]['Y'],G.nodes[point]['X']))
        node_list.remove(point)
        mdis = 999999999999999999999999999999
        for j in node_list:
            dis = 0
            for k in patch_10:
                dis+=(k[0]-G.nodes[j]['Y'])**2 + (k[1]-G.nodes[j]['X'])**2
            if dis<mdis:
                point = j
                mdis = dis

    patch_20 = patch_10[:]

    while pop<20000:
        pop+=G.nodes[point]['NP']
        patch_20.append((G.nodes[point]['Y'],G.nodes[point]['X']))
        node_list.remove(point)
        mdis = 999999999999999999999999999999
        for j in node_list:
            dis = 0
            for k in patch_20:
                dis+=(k[0]-G.nodes[j]['Y'])**2 + (k[1]-G.nodes[j]['X'])**2
            if dis<mdis:
                point = j
                mdis = dis
    
    return patch_5, patch_10, patch_20
```

```python
patch_5, patch_10, patch_20 = create_random_patch(G)
```

```python
def gen_data(patch, dates, G, Cmat, mapping):
    hull = [patch[j] for j in ConvexHull(patch).vertices]
    poly = Polygon(hull)
    pop = 0
    ts = []
    keys = sorted(dates.keys())
    count = 0
    for k in keys:
        for case in dates[k]:
            point = Point(*case)
            if poly.buffer(0.0001).contains(point):
                count+=1
        ts.append((k, count))
    nodes = []
    for j in G:
        point = Point(G.nodes[j]['Y'],G.nodes[j]['X'])
        if poly.buffer(0.0001).contains(point):
            pop+=G.nodes[j]['NP']
            nodes.append(j)

    coeff = 0
    for node in nodes:
        for jnode in nodes:
            coeff+=np.exp(-Cmat[mapping[node]][mapping[jnode]])*G.nodes[node]['NP']*G.nodes[jnode]['NP']
    coeff = float(coeff)/(pop**2)

    return {'timeseries':ts, 'F(S)':coeff, 'area':poly.area, 'total_pop':pop, 'nodes':nodes, 'patch':patch}
```

```python
while True:
    patch_5, patch_10, patch_20 = create_random_patch(G)
    S5 = gen_data(patch_5, dates, G, Cmat, mapping)
    S10 = gen_data(patch_10, dates, G, Cmat, mapping)
    S20 = gen_data(patch_20, dates, G, Cmat, mapping)
    if S5['timeseries'][-1][1]>0:
        break
```

```python
# Below section for creating graphs and monte carlo simul

def monte(N, initI,theta_E, theta_I ,mf, checkpoints, G = G_Distancing):
    beta = 0.185
    sigma = 1/5.2
    gamma = 1/15
    mu_I = 0.004
    p = 0.6
    Q = None
    beta_D = 0.155
    sigma_D = 1/5.2
    mu_D = 0.0004
    theta_E = theta_E
    theta_I = theta_I
    phi_E = 0.4
    phi_I = 0.8
    initN = N
    initI = initI
    initE = initI*4
    initD_E = 0.0
    initD_I = 0
    initR = 0
    initF = 0
    x_i = 0.001
    psi_E = 0.04
    psi_I = 1
    q = 0.3
    gamma_D = gamma

    # Creating dataframe
    time_index=pd.date_range(start='2020-04-01', periods=90)
    df_I=pd.DataFrame(index=time_index)
    df_DI=pd.DataFrame(index=time_index)
    df_R = pd.DataFrame(index=time_index)

    for i in range(1, 40):
        model = SEIRSNetworkModel(G = G,Q = G_Quarantine, beta = beta, sigma = sigma, gamma =gamma, mu_I = mu_I, beta_D = beta_D, sigma_D = sigma_D, 
                          gamma_D = gamma_D, mu_D = mu_D, theta_E = theta_E, theta_I = theta_I, phi_E = phi_E, phi_I = phi_I,
                          psi_E = psi_E, psi_I = psi_I, q = 0.3, initI = initI, initE = initE,initD_E = initD_E, 
                          initD_I = initD_I)



        model.run(T=120, checkpoints=checkpoints,verbose = True)
        time=pd.to_datetime(pd.Series(model.tseries),unit='d',origin='2020-04-01')

        df_tempI=pd.DataFrame({'t':time, i:model.numI})
        df_tempI.set_index('t',inplace=True)
        df_tempI['t']=df_tempI.index.date
        df_tempI.reset_index(drop=True,inplace=True)
        df_tempI.set_index('t',inplace=True)
        df_tempI.reset_index(inplace=True)
        df_tempI = pd.pivot_table(df_tempI, index='t', aggfunc='mean')

        df_I=df_I.merge(df_tempI, how='outer', right_index=True, left_index=True).fillna(np.NaN)


        df_tempDI=pd.DataFrame({'t':time, i:model.numD_I})
        df_tempDI.set_index('t',inplace=True)
        df_tempDI['t']=df_tempDI.index.date
        df_tempDI.reset_index(drop=True,inplace=True)
        df_tempDI.set_index('t',inplace=True)
        df_tempDI.reset_index(inplace=True)
        df_tempDI= pd.pivot_table(df_tempDI, index='t', aggfunc='mean')
        

        df_DI=df_DI.merge(df_tempDI, how='outer', right_index=True, left_index=True).fillna(np.NaN)
        
        df_tempR=pd.DataFrame({'t':time, i:model.numR})
        df_tempR.set_index('t',inplace=True)
        df_tempR['t']=df_tempR.index.date
        df_tempR.reset_index(drop=True,inplace=True)
        df_tempR.set_index('t',inplace=True)
        df_tempR.reset_index(inplace=True)
        df_tempR= pd.pivot_table(df_tempR, index='t', aggfunc='mean')
        

        df_R=df_R.merge(df_tempR, how='outer', right_index=True, left_index=True).fillna(np.NaN)
        
        
        df_I =df_I.interpolate(method='linear', limit_direction='forward', axis=0)
        df_DI =df_DI.interpolate(method='linear', limit_direction='forward', axis=0)
        df_R = df_R.interpolate(method='linear',limit_direction='forward',axis = 0)
        
        # Changes for looking at actual numbers
        df_Iavg=pd.DataFrame.mean(df_I, axis=1)
        df_Iavg = df_Iavg
        df_DIavg=pd.DataFrame.mean(df_DI, axis=1)
        df_DIavg = df_DIavg
        df_Ravg = pd.DataFrame.mean(df_R, axis = 1)
        df_Ravg = df_Ravg
        
        
    return {'df_DI':df_DIavg, 'df_I': df_Iavg,'df_R': df_Ravg}
```
